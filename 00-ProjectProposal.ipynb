## Background and Prior Work

Competitive online video games such as Valorant, a team-based first-person-shooter, provide a natural setting to study how reward and failure influence persistence. Wins and losses are not merely outcomes of play; they engage reinforcement learning mechanisms supported by dopaminergic reward signaling. A common misconception of dopamine is that dopamine neurons always activate upon receiving positive rewards. However, research in Reward Prediction Error (RPE) shows that dopamine neurons don't just respond to any reward: they respond only to rewards that differ from their prediction.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Take food, for example: you realize the dining hall just added your favourite food to the menu, so you ecstatically purchase it. Perhaps the next day, you'll get it again. But if you continue to buy it every single day, for 3 meals a day, for an entire month—you'll probably be sick of it, despite it having been your favourite food. The same applies in games—if you only ever win, at some point you'll get sick of it and quit. 

RPE suggests that unexpected outcomes can shape motivation and prolong engagement, but that still requires the right amount of variability in reinforcement schedules. Just as winning too much can dull your gameplay experience, losing too much could deter you from wanting to experience more. Most players believe that competitive MMR (Matchmaking Rating) systems aim for players to have roughly a 50% win/loss ratio, as "it means it's where you should be," (in terms of rank and skill) but some have also noted that it also creates negative feelings caused by being "hardstuck," <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). a phenomenon where players reach a bottleneck in terms of win/loss ratio, resulting in stagnant rankings that can contribute to feelings of demotivation. Hence, there is reason to suspect that MMR systems are not so transparently "50/50."

Prior works show that there is a temporal attractiveness in uncertainty and resolution—uncertainty is worth it because resolution awaits<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). Our desire to move from a state of "unknown" to "knowing" is in itself valuable, regardless of whether the outcome is positive or negative. This added value of uncertainty is coined "resolution utility," (Ruan et al., 2018, Shen et al., 2019) <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3), and is crucial to why player's don't "quit halfway." The result of this research not only provides reasons for why continuous gameplay occurs, it also helps validate that a player's decision to stop playing is determined by what happens after resolution, rather than issues arising during gameplay.

We specifically chose competitive Valorant because the added layer of social rewards in tandem with wins/losses will provide a stronger reinforcement effect. Competitive Valorant segregates players by skill, resulting in increasing game difficulty as the player improves, thereby ensuring that reward outcomes continue to be variable in nature<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). The rank labels displayed on player accounts introduce social status and evaluation, which can increase motivation to continue playing and make the decision to stop a more consequential choice. 

We recognize that ending a gaming session may come from a variety of reasons: fatigue, obligations (e.g., work or school), social factors, technical interruptions, and more. However, this project focuses on whether match outcome patterns have a statistically meaningful relationship with variation in play duration, which, if found, would indicate cause beyond these background factors.

## Data

Our dataset would require match timestamps to determine consecutive gameplay (which would act as out independent variable), and match outcomes to evaluate win/loss streak effects (our dependent variable). In an ideal dataset, we would also have various control variables including player rank (to control for skill variation), game mode (ranked vs. casual), match duration (to account for fatigue and time constraints), and party information (playing with friends vs. alone). These factors may influence both outcome and persistence, so controlling them would increase the validity of our analysis.

Because win/loss patterns occur over repeated matches, in the most ideal situation, we would collect thousands of matches across many players. For example, a dataset of ~50–100 players with several hundred matches each (10,000+ total matches) would provide enough statistical power to examine session-level relationships and variability across individuals. 

Our goal is to collect data through official game analytics API rather than web-scraping (which may pose ethical concerns). This should provide us with the necessary amount of structure and detail for our analysis. However, we want to note that achieving a large dataset through this method may be difficult. Match history is only accessible by direct-search, meaning we would need consent from individual players to access their data. 

The data would be stored in a tidy, match-level format where each row represents one match played by one player. Columns would include player identifier, timestamp, win/loss outcome, match duration, and control variables such as rank or mode. Session-level features (e.g. streak length, session duration, matches per session) could then be derived during preprocessing and stored in a secondary aggregated dataset.
